```python
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([my_paragraph])
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []

for sentence in my_paragraph.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1, len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from keras.preprocessing.sequence import pad_sequences
pad_input_sequence = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
pad_input_sequence
```

```python
x = pad_input_sequence[:,:-1]
y = pad_input_sequence[:, -1]
```

```python
x.shape
```

```python
y.shape
```

```python
from keras.utils import to_categorical
y = to_categorical(y , num_classes=200)
```

```python
y.shape
```

# Building LSTM Model

```python
from keras import Sequential
from keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(200, 100, input_length=25))
model.add(LSTM(150))
model.add(Dense(200, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(x, y, epochs=100)
```

```python
# predict the next word

import numpy as np
import time
text ="Placement"

for i in range(15):
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=25, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word, index in tokenizer.word_index.items():
    if index==pos:
      text = text +" " + word
      print(text)
      time.sleep(2)
```
